```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("data.csv")
# To display the top 5 rows
df.head(5)
```

```python
df.tail(5)
```

```python
df.dtypes
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "Tenaga Kuda","Year":"Tahun","Make":"Merk","Engine Cylinders": "Silinder", "Transmission Type": "Transmisi", "Driven_Wheels": "Roda","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Harga" })
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()      # Used to count the number of rows
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
sns.boxplot(x=df['Harga'])
```

```python
sns.boxplot(x=df['Tenaga Kuda'])
```

```python
sns.boxplot(x=df['Silinder'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
df.Merk.value_counts().nlargest(40).plot(kind='bar', figsize=(15,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Merk');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="coolwarm",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Tenaga Kuda'], df['Harga'])
ax.set_xlabel('Tenaga Kuda')
ax.set_ylabel('Harga')
plt.show()
```
